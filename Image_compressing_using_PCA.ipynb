```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Load image and convert to grayscale
img = plt.imread('2DtfhLWU_2x.jpg')
gray_img = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
# Flatten the image into a 1D array
n_samples = gray_img.shape[0] * gray_img.shape[1]
X = gray_img.reshape((n_samples, -1))
```

```python
# Perform PCA with 100 principal components
n_components = 1
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)
```

```python
# Reconstruct the image using the principal components
X_reconstructed = pca.inverse_transform(X_pca)
compressed_img = X_reconstructed.reshape(gray_img.shape)
```

```python
# Display the original and compressed images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(gray_img, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(compressed_img, cmap='gray')
ax[1].set_title('Compressed Image with PCA')
plt.show()
```
